## 3. Escribe una funci贸n que sume todos los elementos de una lista

```python
def sum_all(lst):
    """
    USO: Sumar todos los elementos de una lista
    PARAMETROS:
        lst [Lista]: Lista para ser sumada
    VARIABLES:
        suma [int]: Armazena los elementos sumados
    DEVOLUCIN: Devuelve un numero unico con el total de la suma de todos los elementos de la lista.
    """
    suma = 0
    for elemento in lst:
        suma += elemento
    return suma
```

```python
Resultado = sum_all([2,7,4,1,8,5,2,9,6,3])
print(Resultado)
```

## 6. Escribe una funci贸n que devuelva el factorial de un n煤mero.

```python
def factorial(n):
    """
    USO: Calcular o valor factorial de un numero n.
    PARAMETROS:
        n [int]: Numero entero para saber el factorial
    VARIABLES:
        x [int]: Armazena los valores calculados
    DEVOLUCIN: La funci贸n devuelve el valor final de x, que ser谩 el factorial del n煤mero de entrada "n"
    """
    x = 1
    for k in range(1,n+1):
        if n > 0:
            x *= k
        elif n == 0:
            x = 1
        else:
            print("Numero invalido")
    return(x)
```

```python
def unique(lst_un):
    """
    USO: Borrar numero duplicados de una lista
    PARAMETROS:
        lst_un [list]: lista original
    VARIABLES:
        x [list]: Lista vacia para armazenar los valores calculados
    DEVOLUCIN: La funci贸n devuelve la lista con valores unicos.
    """   
    x = []
    for k in lst_un:
        if k not in x:
            x.append(k)
    else:
        pass
    return x
```
